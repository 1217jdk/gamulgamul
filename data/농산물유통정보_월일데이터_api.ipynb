### (2) 배추 & 무 & 귤 product 합치기 & 깐마늘(국산) 바꾸기

````python
### 상품이름 변환
baechu = pdf[pdf['product']=='배추']
baechu = baechu.replace(['봄','월동','고랭지','가을'],'배추')  # 상품 이름을 배추로 통일시키기


mu = pdf[pdf['product']=='무']
mu = mu.replace(['봄','월동','고랭지','가을'],'무')            # 상품 이름을 무로 통일시키기


gyul = pdf[pdf['product']=='감귤']
gyul = gyul.replace(['시설','노지'],'감귤')                   # 상품 이름을 감귤로 통일시키기
````

````python
### groupby로 평균내기
baechu_group = baechu.groupby(['category', 'product','unit','measure', 'weight', 'img',  'goods', '등급', 'date_type','research_date']).mean()
mu_group     =     mu.groupby(['category', 'product','unit','measure', 'weight', 'img',  'goods', '등급', 'date_type','research_date']).mean()
gyul_group = gyul.groupby(['category', 'product','unit','measure', 'weight', 'img',  'goods', '등급', 'date_type','research_date']).mean()


### multiindex 접근
tmp = baechu_group.xs('상품', level='등급')
tmp.info()

### multiindex 삭제
baechu_group = baechu_group.reset_index()
mu_group = mu_group.reset_index()
gyul_group = gyul_group.reset_index()
mu_group
baechu_group
````

````python
pdf['product'] = pdf['product'].replace('깐마늘(국산)','깐마늘')
pdf[pdf['product'] == '깐마늘']
````

### (6) null 처리

````python
### nan 여부 확인
np.isnan(pdf.loc[6947,'price'])
````

````python
pdf['rep_price'] = pdf['price']
````

````python
import numpy as np
for i in range(len(pdf)):
    # print(i)
    cur = pdf.loc[i,'rep_price']
    # 1. 현재가 nan인 경우
    if np.isnan(cur):
        ## [1]  만약 df의 맨 앞이라면 or df맨 앞 제외한, 각 상품의 맨 앞인 경우
        if i == 0 or (pdf.loc[i-1,'product'] != pdf.loc[i,'product']):
            # nan이 아닌 가장 가까운 값가져와서 다 채우기
            j = 1
            while np.isnan(pdf.loc[i+j,'rep_price']):
         
                j += 1
            
            replace = pdf.loc[i+j,'rep_price']
            # 대체하기
            j -= 1
            while j >= 0 :
                pdf.loc[i+j,'rep_price'] = replace
                j -= 1
        
        ## [2] 각 상품의 맨 앞이 아닌 경우 -> 바로 앞의 값으로 대체
        else:
            pdf.loc[i,'rep_price'] = pdf.loc[i-1,'rep_price'] 
            

        
````
