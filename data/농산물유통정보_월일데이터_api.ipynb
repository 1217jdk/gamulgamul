````python
# 검색 API 이용해서, 농산물 검색
import os
import sys
import urllib.request
import pprint, json
p_productclscode = '01'       # 소/도매 (01/02)
p_startday = '2022-03-08'   
p_endday = '2022-09-09'
p_itemcategorycode = '400'   # 품목그룹코드 : 식량작물(100), 채소류(200), 특용작물(300), 과일류(400)
p_itemcode = '429'           # 품목코드 : 쌀, 찹쌀, 고구마 등을 지칭
p_kindcode = '00'            # 픔종코드
p_productrankcode = '04'     # 등급코드 : 상품(04), 중품(05)

p_returntype = 'json'
action='periodProductList'

url = f"http://www.kamis.or.kr/service/price/xml.do?\
action={action}&\
p_productclscode={p_productclscode}&\
p_startday={p_startday}&\
p_endday={p_endday}&\
p_itemcategorycode={p_itemcategorycode}&\
p_itemcode={p_itemcode}&\
p_kindcode={p_kindcode}&\
p_productrankcode={p_productrankcode}&\
p_cert_key={p_cert_key}&\
p_cert_id={p_cert_id}&\
p_returntype={p_returntype}"

# print(url)

request = urllib.request.Request(url)
response = urllib.request.urlopen(request)
rescode = response.getcode()
if(rescode==200):
    response_body = response.read()
    tmp = json.loads(response_body.decode('utf-8'))
    pprint.pprint(tmp)
else:
    print("Error Code:" + rescode)
````

## [1] API
+ 아래 페이지와 같은 결과물로 시장+대형마트 가격평균\
https://www.kamis.or.kr/customer/price/retail/period.do?action=monthly&yyyy=2022&period=3&countycode=&itemcategorycode=100&itemcode=111&kindcode=&productrankcode=0&convert_kg_yn=N

````python
# 검색 API 이용해서, 농산물 검색

import os
import sys
import urllib.request
import pprint, json

p_yyyy = '2022'  # 시작년도
p_period='10'   # 시작년도 기준 과거로 n년까지 가져오기

p_itemcategorycode = '100'   # 품목그룹코드 : 식량작물(100), 채소류(200), 특용작물(300), 과일류(400)
p_itemcode = '111'           # 품목코드 : 쌀, 찹쌀, 고구마 등을 지칭
p_kindcode = '01'            # 픔종코드
p_graderank = '1'     # 등급코드 : 상품(1) , 중품(2)

p_returntype = 'json'
action='monthlySalesList'

url = f"http://www.kamis.or.kr/service/price/xml.do?\
action={action}&\
p_productclscode={p_productclscode}&\
p_yyyy={p_yyyy}&\
p_period={p_period}&\
p_itemcategorycode={p_itemcategorycode}&\
p_itemcode={p_itemcode}&\
p_kindcode={p_kindcode}&\
p_graderank={p_graderank}&\
p_cert_key={p_cert_key}&\
p_cert_id={p_cert_id}&\
p_returntype={p_returntype}"

# print(url)

request = urllib.request.Request(url)
response = urllib.request.urlopen(request)
rescode = response.getcode()
if(rescode==200):
    response_body = response.read()
    tmp = json.loads(response_body.decode('utf-8'))
    pprint.pprint(tmp)
else:
    print("Error Code:" + rescode)
````
