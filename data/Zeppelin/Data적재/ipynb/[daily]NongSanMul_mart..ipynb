```scala
%pyspark
# 1. 날짜 설정
# today = datetime.datetime.now().strftime('%Y-%m-%d')
research_date = ( datetime.datetime.now()  + datetime.timedelta(hours=9) ).strftime("%Y-%m-%d")
research_date = '2022-10-05'

# p_endday = '2022-09-30'
# filedate = '20220930'



# 2. DF 생성
## [1] error 넣을 df 생성
error_pdf = pd.DataFrame(columns=['message','p_itemcategorycode', 'p_itemcode', 'p_kindcode', 'p_productrankcode'])
error_pdf_length = 0


p_productclscode = '01'       # 소/도매 (01/02)
p_returntype = 'json'
action='periodProductList'
p_convert_kg_yn='N'


## [2] 저장할 df 생성
result_pdf = pd.DataFrame(columns=['itemname',	'kindname', 'research_date', 'price','unit_price'])

import pprint


# 3. 데이터 받아오기
for p_itemcategorycode, p_itemcode, p_kindcode, p_productrankcode, p_graderank in api_codes:

    url = f"http://www.kamis.or.kr/service/price/xml.do?action={action}&p_productclscode={p_productclscode}&p_startday={research_date}&p_endday={research_date}&p_itemcategorycode={p_itemcategorycode}&p_itemcode={p_itemcode}&p_kindcode={p_kindcode}&p_productrankcode={p_productrankcode}&p_cert_key={p_cert_key}&p_cert_id={p_cert_id}&p_returntype={p_returntype}&p_countrycode=1101"


    request = urllib.request.Request(url)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode==200):
        response_body = response.read()
        tmp = json.loads(response_body.decode('utf-8'))
        pprint.pprint(tmp)
        # df 생성
        if tmp['data']!=['001']:
            test = pd.DataFrame(tmp['data']['item'])
    
            # 유통만 가져오기
            marketname = list(map(str, test['marketname']))
            test['marketname'] = marketname
            test = test[test['marketname'].str.contains('유통')].reset_index(drop=True)
            
            # 유통 없으면, 그만하기
            if len(test) == 0:
                continue

            
            # 날짜 컬럼 추가
            research_date_col = [ test.loc[i,'yyyy'] + '-' + test.loc[i,'regday'][:2] + '-' + test.loc[i,'regday'][3:5]   for i in range(len(test))]
            test['research_date'] = research_date_col
            
            # 가격 바꾸기
            price = list(test['price'])
            for i in range(len(price)):
                p = price[i]
                if p == '-':
                    price[i] = np.nan
                else:
                    price[i] = int(p.replace(',',''))

            # price = [ int(string_price.replace(',','')) for string_price in price    ]
            test['price'] = price
        
            # groupby
            test = test.groupby(['itemname','kindname','research_date']).aggregate({'price':'mean'})
            test.round({'price':2})
            test = test.reset_index()
    
            # unit_price 추가 : 여기는 row 1개 짜리 pandas DF 임
            if test.loc[0,'itemname']=='쌀':
                test['unit_price'] = test['price']/20
            
            elif '팽이버섯' in test.loc[0,'itemname']:
                test['unit_price'] = test['price']*(2/3)
            
            else:
                test['unit_price'] = test['price']
        
            # concat
            result_pdf = pd.concat([result_pdf, test])
        
        
        # 데이터 없거나, 에러인 경우
        else:
            error_pdf.loc[error_pdf_length] = [tmp['data'][0], p_itemcategorycode, p_itemcode, p_kindcode, p_productrankcode ]
            error_pdf_length += 1
    else:
        print("Error Code:" + rescode)

```
