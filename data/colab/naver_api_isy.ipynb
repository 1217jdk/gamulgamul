```python
def get_goods_info(mallName, goods_name):
  encText = urllib.parse.quote(mallName+goods_name)
  url = "https://openapi.naver.com/v1/search/shop.json?exclude=used:rental:cbshop&start=1&display=30&sort=sim&query=" + encText 
  request = urllib.request.Request(url)
  request.add_header("X-Naver-Client-Id",client_id)
  request.add_header("X-Naver-Client-Secret",client_secret)
  response = urllib.request.urlopen(request)
  rescode = response.getcode()

  if(rescode==200):
    response_body = response.read()
    res = json.loads(response_body.decode('utf-8'))
    # print(response_body.decode('utf-8'))
  else:
    print("Error Code:" + rescode)
    return
  goods_info = dict()
  goods_info['goods_name'] = goods_name
  for i in range(res['display']):
    if mallName in res['items'][i].get('mallName').replace(' ',''):
      goods_info['goods_price'] = res['items'][i].get('lprice')
      goods_info['goods_date'] = datetime.datetime.now().strftime('%Y-%m-%d')
      break
  goods_info['business'] = 'm'
  
  return goods_info
```

```python
def get_goods_cheap(goods_name):
  encText = urllib.parse.quote(goods_name)
  url = "https://openapi.naver.com/v1/search/shop.json?exclude=used:rental:cbshop&start=1&display=30&sort=sim&query=" + encText 
  request = urllib.request.Request(url)
  request.add_header("X-Naver-Client-Id",client_id)
  request.add_header("X-Naver-Client-Secret",client_secret)
  response = urllib.request.urlopen(request)
  rescode = response.getcode()

  if(rescode==200):
    response_body = response.read()
    res = json.loads(response_body.decode('utf-8'))
    # print(response_body.decode('utf-8'))
  else:
    print("Error Code:" + rescode)
    return
    
  cheap_info = dict()
  if not res['items']:
    return f'{goods_name}_img', f'{goods_name}_url', {'goods_name': goods_name, 'cheap_price': None, 'cheap_date': datetime.datetime.now().strftime('%Y-%m-%d'), 'business': 'o'} 
  goods_img = res['items'][0].get('image')
  cheap = sorted(res['items'], key= lambda x: int(x['lprice']))[0]
  cheap_url = cheap.get('link')
  cheap_info['goods_name'] = goods_name
  cheap_info['cheap_price'] = cheap.get('lprice')
  cheap_info['cheap_date'] = datetime.datetime.now().strftime('%Y-%m-%d')
  cheap_info['business'] = 'o'
  
  return goods_img, cheap_url, cheap_info
```
