## 주차별 모두 합치기

```python
tprice_l = tprice_l_1.copy()
tprice_l.drop(columns=['바코드'], inplace=True)
tprice_l['총용량'] = tprice_l['총용량'].astype(str)
tprice_l['단위용량'] = tprice_l['단위용량'].astype(str)
for i in range(2, 7):
  tmp_df = pd.read_excel(f'참가격_슈퍼마켓_롯데슈퍼_{i}.xlsx', usecols=range(1, 22))
  tmp_df.drop(columns=['바코드'], inplace=True)
  tmp_df['총용량'] = tmp_df['총용량'].astype(str)
  tmp_df['단위용량'] = tmp_df['단위용량'].astype(str)
  tprice_l = pd.merge(left=tprice_l, right=tmp_df, how='outer', on=['판매업소명', '업태명', '업소체인명', '대분류명', '중분류명', '소분류명', '상품명', '제조사', '총용량', '총용량명', '단위용량', '단위용량명'])
```

## 행, 열 정리 및 '/', 결측치 처리

```python
agr_idx = tprice_l[tprice_l['중분류명'] == '채소류'].index
tprice_l.drop(agr_idx, inplace=True)
```

```python
import numpy as np
tprice_l.drop(columns=['판매업소명', '대분류명', '중분류명', '제조사',], inplace=True)
tprice_l = tprice_l.replace('-', np.NaN)
```

```python
tprice_l.head()
```

```python
for i in range(8, 54):
  for j in range(len(tprice_l)):
    if pd.isna(tprice_l.iloc[j, i]):
      if i == 8:
        for k in range(i+1, 54):
          if not pd.isna(tprice_l.iloc[j, k]):
            tprice_l.iloc[j, i] = tprice_l.iloc[j, k].split('/')[1]
            break
      else:
        tprice_l.iloc[j, i] = tprice_l.iloc[j, i-1]
    else:
      tprice_l.iloc[j, i] = tprice_l.iloc[j, i].split('/')[1]
```

```python
df_basic = tprice_ssm.drop(tprice_ssm.iloc[:, 6:52], axis=1)
df_basic.drop_duplicates(['name'], ignore_index=True, inplace=True)
df_basic
```

```python
dates = ['20210910', '20210917', '20210924', '20211001', '20211008', '20211015', '20211022',
       '20211029', '20211105', '20211112', '20211119', '20211126', '20211203',
       '20211210', '20211217', '20211224', '20211231', '20220107', '20220114',
       '20220121', '20220128', '20220204', '20220211', '20220218', '20220225',
       '20220304', '20220311', '20220318', '20220325', '20220401', '20220408',
       '20220422', '20220429', '20220506', '20220513', '20220520', '20220527',
       '20220603', '20220610', '20220708', '20220715', '20220722', '20220729',
       '20220805', '20220812', '20220819']
```

```python
# ssm 상품별 롯데슈퍼+GS더프레시 평균 가격 데이터
df_list = []
for date in dates:
  d = date[:4] + '-' + date[4:6] + '-' + date[6:]
  tmp_df = df_basic.copy()
  tmp_df['research_date'] = d
  tprice_ssm[date] = tprice_ssm[date].astype(int)
  mean_tmp = tprice_ssm.groupby('name')[date].mean().round(2)
  mean_list = []
  for i in range(len(tmp_df)):
    mean_list.append(mean_tmp[tmp_df.iloc[i, 3]])
  tmp_df['price'] = mean_list
  df_list.append(tmp_df)

df_ssm = pd.concat(df_list, ignore_index=True)
```
