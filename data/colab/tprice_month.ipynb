# 전처리 방안

## [1] null 값 전처리 방법
1. 앞의 값으로 뒤의 null 값 채우기
2. 앞 뒤의 값의 평균으로 null 값 채우기

## [2] 전처리 단위 데이터
1. 연데이터 마다 null 값 채운 후, 10년치 합치기
    + 장점
        + 우리가 인식할 수 있는 범위내에서, null처리를 명확하게 할 수 있음
    + 단점
        + 그렇게 되면 2012년와 2022년과 row개수 달라서, 또 한번 더 null 처리 해줘야 함
        + 2012-01 데이터와 2013-01 데이터가 있고 나머지가 null일 때, 2012-12의 대체값을 2012-01로 쓰게 됨

2. 2012~2022년 데이터 모두 모아서 한 번에 null 처리하기
    + 장점
        + 한 번만 null처리 해주면 됨

    + 단점
        + 과거의 데이터가 없는 상품들에 대해서, 과거부분 null 처리를 할지 말지를 정해야함
        + 우리가 인식하지 못하는 범위에서, 너무 많은 부분의 null들이 대체값으로 처리될 수 있음

## 가격 '/' 처리 & null('-' 처리)
'/' 기준으로 split 0번째\
null 처리는 이전 값이 있으면 다음 값을 똑같은 값으로 채워주는데\
만약 첫번째 값부터 null이라면 이후 값들 중 처음으로 null이 아닌 값을 채워줌

```python
print(tprice_2012.iloc[0, 14])
```

```python
for i in range(4, 16):
  for j in range(len(tprice_2012)):
    # print(tprice_2012[i][j].split('/')[0])
    tprice_2012.iloc[j, i] = tprice_2012.iloc[j, i].split('/')[0]
    if tprice_2012.iloc[j, i] == '-':
      if i == 4:
        for k in range(i+1, 16):
          if tprice_2012.iloc[j, k] != '-':
            tprice_2012.iloc[j, i] = tprice_2012.iloc[j, k].split('/')[0]
            break
      else:
        tprice_2012.iloc[j, i] = tprice_2012.iloc[j, i-1]
      # tprice_2012[i][j] = None
```
